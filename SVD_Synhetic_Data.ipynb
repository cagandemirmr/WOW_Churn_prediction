```python
from sdv.metadata import Metadata
from sdv.single_table import GaussianCopulaSynthesizer
import pandas as pd



# Metadata nesnesini oluştur
metadata = Metadata.detect_from_dataframe(
    data=df)
metadata.save_to_json(filepath='/content/sample_data/my_metadata_v1.json') # Create Json File for future usages


```

```python
from sdv.single_table import GaussianCopulaSynthesizer #Creating Model

synthesizer = GaussianCopulaSynthesizer(metadata)
synthesizer.fit(data=df)
```

```python
synthetic_data = synthesizer.sample(num_rows=20000)
```

```python
from sdv.evaluation.single_table import evaluate_quality

quality_report = evaluate_quality(
    df,
    synthetic_data,
    metadata)
```

```python
synthetic_data.head()
```

```python
synthetic_data = synthesizer.sample(num_rows=2000)
quality_report = evaluate_quality(
    df,
    synthetic_data,
    metadata)
```

```python
synthetic_data.to_csv('synthetic_data.csv', index=False)
```
